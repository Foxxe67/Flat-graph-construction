```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def get_graph_from_filr(filename):
    myGraph = nx.Graph()
    f = open(filename, 'r')
    for line in f.readlines():
        for reb in line.split(' '):
            if reb != '-1\n' and reb != '-1':
                first = reb.split('-')[0]
                second = reb.split('-')[1]
                myGraph.add_edge(first,second)
                f.close()
    return myGraph

nx.draw(get_graph_from_filr('test1.txt'),with_labels=True)
```
